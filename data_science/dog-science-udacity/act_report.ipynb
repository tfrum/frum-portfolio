# Insights

Overall this dataset contains a few unsurprising conclusions and features. This was a bit of a blowout for breeds like Golden Retrievers and Labradors, which saw heavy representation in the distribution of dog breeds. The data for dog breeds is hardly normal. There are a few breeds with a lot of references, and a whole lot with maybe one. It should be noted that man of those single number "breeds" aren't breeds, as we didn't filter our count on the untrustable boolean columns which identify whether or not a prediction item is a dog or not. We can't trust it, so we just didn't use it. Our top results are all dogs, anyway.

After that we wanted to look at the scores themselves. We did make some effort to clean them up and drop outliers, but the metric isn't clean enough to do that quickly. Even after some cleaning, we were left with some issues. However, the data is surprisingly normal and its biases are clearly defined in a simple histogram.

What we can see from the resulting graph is that the data is normal-ish. This tells us there's some validity to the scores used in this measure. However, they're right-skewed where they are distributed, and that implies that scores are generally higher than lower.

Likewise, scores are above 1, which means that a majority of scores had numerators higher than denominators, such as 11/10. We also didn't catch all of our edgecases, which is shown by the outliers on the right-end of the data.

Finally, that pesky boolean metric needed to be addressed. I don't think it's particularly accurate for certain breeds, so we didn't take it into account here. However, they were mostly consistent in how often a non-breed object was used across the three options. This tells us there's at least some method to the madness of those "prediction" columns.
